## REDDIT API

```python
import praw
import re
```

```python
reddit = praw.Reddit(
    client_id="#######################",
    client_secret="########################",
    user_agent="<console:FreeStuff:1.0>",
)

subreddit = reddit.subreddit('FreeGameFindings')

# OPEN FILE TO CHECK WHAT WAS LAST GAME
f = open('last_game.txt','r')
last_game = f.read()

# ---- VARIABLES
# NEW LAST GAME TO BE APPENDED TO TXT
new_last_game = ''
# MAIN LIST TO BE RETURNED
reddit_list = []



# STEAM....
def get_platform(title):
    pattern = r"\[.*?\]"
    platform = re.findall(pattern,title)
    platform = platform[0][1:-1]
    return platform

# FAR CRY 3 
def get_game(title):
    t_pat = r"\)(.*)"
    game = re.findall(t_pat,title)
    game = str(game)[2:-2]
    return game

# GAME DLC OTHER
def get_type(title):
    dlc_pat = r"\(.*?\)"
    dlc = re.findall(dlc_pat,title)
    dlc = str(dlc)[3:-3]
    return dlc

def reddit_content_main():
    # Count Index of Posts
    counter = 0
    
    # Empty Temp Dictionary
    temp = {}
    
    for post in subreddit.new(limit=100):
        if post.id == last_game:
            break
        elif not post.is_self:
            if counter == 0:
                new_last_game = post.id
            platform = get_platform(post.title)
            game = get_game(post.title)
            url = post.url
            dlc = get_type(post.title)
            f2 = open('last_game.txt','w')
            f2.write(new_last_game)
            counter += 1
            temp['game'] = game
            temp['platform'] = platform
            temp['url'] = url
            temp['type'] = dlc
            temp_copy = temp.copy()
            reddit_list.append(temp_copy)
    
    # CLEANING REDDIT DATA

    print("REDDIT DONE")
    return reddit_list
                
```

```python
from serpapi import GoogleSearch
import socket
import os
import urllib.request
import time
socket.setdefaulttimeout(15)
```

```python



images_links = []

# FETECH IMAGES FOR GAME
def google_images_links(game_name):
    google_params = {
      "q": "",
      "tbm": "isch",
      "ijn": "0",
      "api_key": "################################"
    }
    
    google_params['q'] = game_name
    
  
    search_obj = GoogleSearch(google_params)
    results = search_obj.get_dict()
    images_list = results['images_results']
    

    for image in images_list:
      images_links.append(image['original'])
    print("IMAGES FETECHED")
    return images_links

# DOWNLOAD IMAGES FOR GAME
def download_google_images(img_num , images_links):
    print("_---------------------------------------")
    for i in range(img_num):
        path = r"C:\\Users\\Parme\\Music\\Untitled Folder\\Gimgs\\"
        try:
            name = "" + str(i) + ".png"
            path = path + name
            urllib.request.urlretrieve(images_links[i], path)
        except Exception as e:
            print("error for ",e)
            pass

# DELETE THE FILES
def delete_temp_images(path):
    for file in os.listdir(path):
        if file.endswith('.png'):
            temp_path = path + file
            os.remove(temp_path)
```

## FIND PERFECT BACKGROUND IMAGE

```python
from PIL import Image, ImageDraw , ImageFont
```

```python

# Resize Image
# Add Alpha
# Paste Template onto the Image

def design_post(template_path , game_img_path, img_name):

    
    template = Image.open(template_path)
 
    try:
        game_img = Image.open(game_img_path)
        game_img = game_img.resize((540,540))
        game_img.putalpha(80)
        game_img.paste(template,(0,0),template)
    except Exception as e:
        print("PASTING ERROR" , e)
        pass
    
    # SAVE EDITED PICTURE
    path = r"C:\Users\Parme\Music\Untitled Folder\Designed" + "\ ".strip()
    name = "Designed" + str(img_name)
    path = path + name + ".png"
    try:
        game_img.save(path)
    except Exception:
        pass
    
```

## ADD TEXT TO IMAGE

```python
import textwrap
import json
```

```python

# Decide Monospace Font
font1 = ImageFont.truetype("MAKISUPA.TTF", 30, encoding="unic")
font2 = ImageFont.truetype("MAKISUPA.TTF", 15, encoding="unic")


# Get width wrap onto Game Title
def wrap_text(text, text_width):
    wrapped_text = textwrap.wrap(text , width = text_width)
    return wrapped_text


# Center TEXT if small
def center_text(word_fontsize_x):
    difference = 260 - word_fontsize_x
    text_x = round(difference / 2)
    # 205 is starting point for text on img
    text_x = 205 + text_x
    return text_x
    

# Add TEXT to Image
def add_text_main(designed_img_path , game_name , platform_name,url, img_name,designed_counter,folder_path):
    try:
        designed_img = Image.open(designed_img_path)
        draw = ImageDraw.Draw(designed_img)
        wrapped_game_name = wrap_text(game_name , 13)
        text_y = 250
        for word in wrapped_game_name:
            game_word_fontsize_x = font1.getsize(word)[0]
            game_word_x = center_text(game_word_fontsize_x)
            draw.text((game_word_x ,text_y) , word , font=font1 , fill = (54,57,112))
            text_y += 40
    
        # PLATFORM
        platform_fontsize_x = font2.getsize(platform_name)[0]
        platform_x = center_text(platform_fontsize_x)
        draw.text((platform_x,425),platform_name,font= font2,fill=(54,57,112))


        path = r"C:\Users\Parme\Music\Untitled Folder\Texted" + "\ ".strip()
        name = str(designed_counter) + "TEXTED" + str(img_name)
        img_path = folder_path + name + ".png"
        designed_img.save(img_path)
        
        # WRITE JSON FILE
        json_data = {
                      "game_name": game_name,
                      "game_platform": platform_name,
                      "game_url": url
                    }
        
        with open('data.json', 'w', encoding='utf-8') as f:
            json.dump(json_data, f, ensure_ascii=False, indent=4)
    
    except:
        pass
```

### SENDING MAIL

```python
import smtplib
from email.message import EmailMessage
import imghdr
```

```python
def mail(attachment_path , game_name , game_platform , hashtags , game_url):

	# PREPARE MAIL BODY
	mail_body = game_name + " is FREE ♥ on " + game_platform + "  "  + hashtags + "  " + game_url

	# MAIL OBJECT
	msg = EmailMessage()
	msg['Subject'] = game_name
	msg['From'] = '#333333333333333'
	msg['To'] = '######################'
	msg.set_content(mail_body)


	# --------location for picture || rb - read bytes || f var
	#to open or read
	with open(attachment_path, 'rb') as f:
		file_data = f.read()
		#returns file type eg. jpg
		file_type = imghdr.what(f.name) 
		# returns file name
		file_name = f.name

	msg.add_attachment(file_data, maintype = 'image', subtype = file_type, filename = file_name)


	with smtplib.SMTP_SSL('smtp.gmail.com', 465) as smtp:
		# id and password
		smtp.login('################', '############')
		smtp.send_message(msg)
		print("Sent")
```
